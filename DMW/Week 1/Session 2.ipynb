JavaScript Object Notation (JSON)
This was developed for JAvaScript. It was designed for a JavaScript object to be serialized. in serializing, you store is as a stream of bytes. When you are sending data through the web, it is actually a series of bytes.

Notice that is has key-value pairs in the syntax of a dictionary, but they are not exactly the same. Notice, for the spouse key, it has ```null```. Instead, you would have used ```None``` in Python. Additionally, all keys are strings. The JSON example below is fairly readable in a human perspective. It has several layers of nesting, and thus it can be verbose. 

When you get a JSON object, you actually just get a string.




````python
json_str = """{ 
  "firstName": "John", 
  "lastName": "Smith", 
  "isAlive": true, 
  "age": 27, 
  "address": { 
    "streetAddress": "21 2nd Street", 
    "city": "New York", 
    "state": "NY", 
    "postalCode": "10021-3100" 
  }, 
  "phoneNumbers": [ 
    { 
      "type": "home", 
      "number": "212 555-1234" 
    }, 
    {
      "type": "office", 
      "number": "646 555-4567" 
    } 
  ], 
  "children": [], 
  "spouse": null 
}
"""

json_str
````

We have operations from the ```json``` library.

```load``` - Read JSON from file object.



````python
# We convert json into nice data types

import json

# We use loads to convert the string into a python dictionary.

json_obj = json.loads(json_str)


print(len(json_obj['children']))

print(json_obj['phoneNumbers'][0]['number'])

# The two lines are the same thing.

[json_obj['phoneNumbers'][i]["number"] for i in range(len(json_obj['phoneNumbers']))]

[phone['number'] for phone in json_obj['phoneNumbers']]
````
