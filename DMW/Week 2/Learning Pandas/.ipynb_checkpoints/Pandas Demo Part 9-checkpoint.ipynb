```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
# dropna is using some default arguments of axis = index and how = any
df.dropna()
```

The axis can be either by index or columns. It would drop columns if they had missing columns. The how is the criteria for dropping a row or column. For any, it drops row for any missing values. If how = all, then it must mean that all of the values are missing values to be dropped.

```python
df.dropna(axis = 'index', how = 'any')
```

```python
df.dropna(axis = 'index', how = 'all')
```

```python
# Since all the columns had some form of missing values, then it all of the columns are just gona
df.dropna(axis = 'columns', how = 'any')
```

```python
# We can pass in the subset argument if we only want to dro specific rows.
# With this notation, for a row to be dropped, as long as either the last name or email is not given, it will be dropped
df.dropna(axis = "index", how = "any", subset=['last', 'email'])
```

```python
# For an index to be dropped, both the last name and email will have to be dropped.
# This is not permanent, so you would need it to be permanent by placing inplace = True
df.dropna(axis = "index", how = "all", subset=['last', 'email'])
```

```python
# We can manually replace cells
df.replace('NA', np.nan, inplace = True)
df.replace('Missing', np.nan, inplace = True)
df
```

```python
df.dropna()
```

```python
# This will give a mask if it is classified as NaN or not.
df.isna()
```

```python
# There also exists to fill the NaN
# This is most useful for numerical data to assign NaN as a value of 0.
df.fillna('MISSING')
```

```python
df.fillna(0)
```

```python
df['age'].describe()
```

```python
df
```

```python
df.dtypes
# notice that even for age, it is an object and NOT an integer
```

```python
# Note that when dealing with numbers, best to use the float since the NaN is a float
type(np.nan)
```

```python
# We will now convert the age from stirng to an integer
df['age'] = df['age'].astype(float)
```

```python
df['age'].describe()
```

```python
# dtypes is an attribute, NOT a method. As such, it does not need a parantheses at the end.
df.dtypes
```
