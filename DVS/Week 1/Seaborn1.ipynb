```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

Always perform your univariate analytics first.

```python
titanic.describe()
```

```python
titanic['survived'].value_counts()
```

We want to do this for all categorical columns in our dataframe.

```python
categorical_columns = titanic.select_dtypes(include =['object','category']).columns.tolist()
```

```python
categorical_columns
```

```python
titanic['survived'] = titanic['survived'].astype('category') 
```

Let us visualize!
histplot and countplot

```python
assert "age" in titanic.columns, "The dataset must contain an 'age' column."
assert not titanic["age"].isnull().all(), "The 'age' column must not be completely empty."
assert titanic["age"].dtype in [float, int], "The 'age' column must contain numeric values."
```

```python
sns.set_theme()
```

```python
sns.histplot(titanic['fare'])
```

```python
sns.countplot(titanic['survived'])
```

Pairwise visualization!

```python
sns.scatterplot(x='age', y='fare', data = titanic)
```

```python
sns.scatterplot(x='age', y='fare', data = titanic, hue = 'pclass')
```

```python
sns.scatterplot(x='sex', y='fare', data = titanic)
```
