```python
# general imports
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load
car = sns.load_dataset("car_crashes")

# White Theme
sns.set_theme(style="white", palette=None)
```

# Design Choice

I went with the white theme because I wanted to avoid the grid lines for the univariates. I didn't find it necessary, because the point of the assignment was to find the central tendency which can be done with just the bars and kde. For that reason, I also picked a lighter color so I could see the kde better.

# Univariate

```python
# Thank you for the idea, JF
schema = {
    'total': 'Total Collisions per Billion Miles',
    'speeding': '% Fatal Crash Drivers Speeding',
    'alcohol': '% Fatal Crash Drivers Alcohol-Impaired',
    'not_distracted': '% Fatal Crash Drivers Not Distracted',
    'no_previous': '% Fatal Crash Drivers Without Prior Accidents',
    'ins_premium': 'Car Insurance Premiums ($)',
    'ins_losses': 'Losses per Insured Driver ($)',
    'abbrev': 'State Abbreviation'
}


schema_df = pd.Series(schema)

def my_graph(column):
    # Graph
    graph = sns.displot(car[column], kde = True)

    # Labels
    plt.xlabel(schema_df[column])
    plt.ylabel('Frequency')

    # Calculations
    mean = car[column].mean()
    median = car[column].median()

    # Central Tendency
    print(f"Mean: {mean}")
    print(f"Median: {median}")
```

## Total Collissions
The total collisions are normally distributed. Because of this, we prefer the average as a measure of central tendency. With this in mind, the graph tells us that, on average, people can expect around 15-16 collisions for every billion miles traveled.

```python
my_graph('total')
```

## Speeding
For this data, we prefer to use the median as the measure of central tendency because the graph is skewed. We know this because the function is densest to the left with outliers trailing to the right. This graph shows us the percentage of drivers who were speeding and involved in fatal collissions. From the median, we expect that around 4.6% of speeding drivers will be involved in a collission for every billion miles traveled.

```python
my_graph('speeding')
```

## Alcohol
The next chart features the percentage of drivers involved in fatal collissions who were alcohol-impaired. We see that the graph is skewed to the right, with some statements reporting a collission rate of 10%. Thus, we prefer to utilize the median. While most states can expect a rate of 4.55%, some states can go as high as 10%.

```python
my_graph('alcohol')
```

## Not Distracted
The data distribution is slightly skewed to the left, but not enough for us to prefer the median. On average, those who aren't distracted while on the road can expect 13.86% chance of encountering a fatal collision every billion miles traveled.

```python
my_graph('not_distracted')
```

## No Previous
The preferred measure of central tendency is the mean because the data is normally distributed. For every 1 billion miles driven, around 14% of drivers who have never had any previous accidents are met with a fatal collision.

```python
my_graph("no_previous")
```

## Insurance 
For this graph, we prefer the median as a measure of central tendency. The graph is skewed to the right. Every 1 billion miles, people can expect to pay insurance premiums of around $858.97.

```python
my_graph("ins_premium")
```

## Insurance Losses
The graph is not skewed, so we prefer the average as a measure of central tendency. Every 1 billion miles driven, insurance companies can expect to lose around $134.49.

```python
my_graph("ins_losses")
```

# Bivariate

```python
def my_relationship(a, b):
    # Graph
    graph = sns.regplot(data=car, x=car[a], y=car[b])
    return graph
```

## Relationship Between Speeding and Alcohol-Impaired
The graph below suggests a linear relationship between the number of fatal collisions caused by speeding and the ones caused by alcohol. It could be that drinking alcohol causes more impulsive behavior, such as speeding.

```python
my_relationship("speeding","alcohol")
```

## Relationship Between Not Distracted and Not Involved
The graph below suggests a linear relationship between the drivers who were not distracted and the drivers who were not previously involved in accidents. It could be that drivers who had no previous accidents paid attention to the road, allowing them to avoid getting into any accident before the unfortunate event.

```python
my_relationship("not_distracted","no_previous")
```
