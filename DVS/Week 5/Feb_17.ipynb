```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

shreks_palette = ['#4a7600', '#FFA500', '#182f7c', '#99303b', '#bf0475']
sns.set_palette(shreks_palette)
sns.palplot(shreks_palette)
plt.show()
```

```python
characters = ['Reader', 'Death', 'Lord Farquadd', 'Shrek', 'Puss in Boots']
num = [121, 697, 135, 1234, 770]
data = pd.DataFrame({'Character': characters, 'Number': num}).sort_values(by = ['Number'], ascending = False)

chart = sns.barplot(x='Character', y='Number', hue = 'Character', data=data, palette=shreks_palette)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.ylabel('Number of Fanfictions ', rotation=0, labelpad=30, ha='left')  # Aligns left
plt.xlabel('')
chart.yaxis.set_label_coords(0, 1.02)
plt.title(f'Popularity of Shrek Characters on A03:\nHow often do Shrek characters appear in Fan Fiction?\n\n', loc='left', fontweight = 'bold')
```
